```python
import numpy as np
from IPython.display import Image
```

```python
def peb_volume(Rp):
    return 4./3 * np.pi * Rp**3
def phi2N(phi, vpeb, vbed):
    return phi* vbed/vpeb
def N2phi(N, vpeb, vbed):
    return N*vpeb/vbed
```

```python
N  = 10
Rp = 0.5
dp = 2*Rp
vp = peb_volume(Rp)
H = 2
vbed = 2*dp * 2*dp * H*dp
phi = 0.62
N2phi(N, vp, vbed)
```

#Looking at pebble beds with thermal cycling

```python
Rp = 0.0005
dp = 2*Rp
vp = peb_volume(Rp)

N = 6660
```

###Initial packing state before thermal cycle

scatter plot of the pebbles projected into the x-s plane

```python
Image(filename='assets/images/x-z-scatter-000000.png') 
```

The packing fraction is found from the average bed height (the z value of the highest 50 pebbles plus a radius)

```python
phii = N2phi(N, vp, 15*dp*15*dp*(0.0255920+0.0005))
```

### Final packing state after 10 thermal cycles

scatter plot of the pebbles projected into the x-z plane

```python
Image(filename='assets/images/x-z-scatter-000299.png') 
```

The final packing fraction is similarly foudn from the average bed height after 10 thermal cycles

```python
phif = N2phi(N, vp, 15*dp*15*dp*(0.024914208+0.0005))
```

### Change in packing

The change in packing fraction is found relative to the initial packing fraction

```python
phidelta = np.abs((phii - phif)/phii)*100

print "initial phi: %s" % phii
print "final phi: %s" % phif
print "change in phi: %s" % phidelta
```
