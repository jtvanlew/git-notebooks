# Stress-strain data from LIGGGHTS

These are all coming from the new routine where $\frac{\rho_\text{scaled}}{\rho} = 1 \times 10^{12}$. This is allowing $\Delta t = 0.1$ which is infinitely better (numerically speaking) than all my previous DEM sims where $\Delta t = 0.000000001$. Quasi-static uniaxial compressions need a very small velocity (and therefore a very long time of compression). Thus larger $\Delta t$ is much preferred.

here $\mu = 0.1$, $E = 90$ GPa. $D_p = 0.5$ mm.

To remove the strain before there is 'good contact' on the pebble bed, I'll subtract all the data from below 0.1 MPa and then shift all the strain data to that point

```julia
data = readtable("FZK-3-stress-strain.csv");
stressFull = data[:Pave_MPa_];
stress0 = stressFull[1:indmax(stressFull)];
stress1 = stressFull[indmax(stressFull)+1:end];
travel = readdlm("FZK-3-travel.txt", '\n');
returnTravel = readdlm("FZK-3-returnTravel.txt", '\n');
stress0 = stress0[end-(length(travel)-1):end];

N = searchsortedfirst(stress0, 0.1)
stress0 = stress0[N:end]
travel  = travel[N:end]
returnTravel = returnTravel - travel[1];
travel  = travel - travel[1]


plot(travel,stress0,color="blue")
plot(returnTravel,stress1,color="blue")






data = readtable("FZK-4-stress-strain.csv");
stressFull = data[:Pave_MPa_];
stress0 = stressFull[1:indmax(stressFull)];
stress1 = stressFull[indmax(stressFull)+1:end];
travel = readdlm("FZK-4-travel.txt", '\n');
returnTravel = readdlm("FZK-4-returnTravel.txt", '\n');
stress0 = stress0[end-(length(travel)-1):end];
N = searchsortedfirst(stress0, 0.1)
stress0 = stress0[N:end]
travel  = travel[N:end]
returnTravel = returnTravel - travel[1];
travel  = travel - travel[1]

plot(travel,stress0,color="green")
plot(returnTravel,stress1,color="green")



data = readtable("FZK-5-stress-strain.csv");
stressFull = data[:Pave_MPa_];
stress0 = stressFull[1:indmax(stressFull)];
stress1 = stressFull[indmax(stressFull)+1:end];
travel = readdlm("FZK-5-travel.txt", '\n');
returnTravel = readdlm("FZK-5-returnTravel.txt", '\n');
stress0 = stress0[end-(length(travel)-1):end];
N = searchsortedfirst(stress0, 0.1)
stress0 = stress0[N:end]
travel  = travel[N:end]
returnTravel = returnTravel - travel[1];
travel  = travel - travel[1]

plot(travel,stress0,color="black")
plot(returnTravel,stress1,color="black")


data = readtable("FZK-6-stress-strain.csv");
stressFull = data[:Pave_MPa_];
stress0 = stressFull[1:indmax(stressFull)];
stress1 = stressFull[indmax(stressFull)+1:end];
travel = readdlm("FZK-6-travel.txt", '\n');
returnTravel = readdlm("FZK-6-returnTravel.txt", '\n');
stress0 = stress0[end-(length(travel)-1):end];
returnTravel = returnTravel[1:length(stress1)];
N = searchsortedfirst(stress0, 0.1)
stress0 = stress0[N:end];
travel  = travel[N:end];
returnTravel = returnTravel - travel[1];
travel  = travel - travel[1];

plot(travel,stress0,color="red")
plot(returnTravel,stress1,color="red")
```
