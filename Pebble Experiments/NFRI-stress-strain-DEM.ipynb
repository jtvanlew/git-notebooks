```julia
using PyPlot;
using DataFrames;
```

# $ v = 1\times10^{-5}$

1-strain_edit folder

```julia
data = readtable("NFRI-1-stress-strain0.csv");
stressFull = data[:P_ave_MPa];
stress0 = stressFull[1:indmax(stressFull)];
stress1 = stressFull[indmax(stressFull)+1:end];
travel = readdlm("NFRI-1-travel.txt", '\n');
returnTravel = readdlm("NFRI-1-returnTravel.txt", '\n');
stress0 = stress0[end-(length(travel)-1):end];
figure(1)
plot(travel,stress0,color="blue")
plot(returnTravel,stress1,color="blue")
```
