```python

def hertz( nu1, E1, nu2, E2, dp, s ):
    # given:
    # Young's Moduli - E in GPa
    # Pebble diameter - dp in mm
    # displacement - s in mm
    dp = dp/1000.
    s = s/1000.
    E1 = E1*10**9.
    E2 = E2*10**9.
    Estar = 1./((1.-nu1**2.)/E1+(1.-nu2**2.)/E2)
    F = (1./3) * Estar * (s**3. * dp)**(1/2.)
    return F


s = linspace(0,0.02,1001) #0.1 mm strain
E = {'Cu':117., 'SS':200., 'Al':69., 'Li4Si04':30}
nu = {'Cu':0.27, 'SS':0.25, 'Al':0.25, 'Li4Si04':0.24}



# Properties
dp = 3.2 #mm
F_n = hertz(nu['Cu'], E['Cu'], nu['SS'], E['SS'], dp, s)
figure(1)
title('Copper')
plot(s, F_n, linewidth = 4)
grid(b=True, which='major', color='b', linestyle='-')
xlim([0, s[-1]])

dp = 4.28
F_n = hertz(nu['Al'], E['Al'], nu['SS'], E['SS'], dp, s)
figure(2)
title('Aluminum')
plot(s, F_n, linewidth = 4)
grid(b=True, which='major', color='b', linestyle='-')
xlim([0, s[-1]])


dp = 0.6
F_n = hertz(nu['Li4Si04'], E['Li4Si04'], nu['SS'], E['SS'], dp, s)
figure(3)
title(r'Li$_4$SiO$_4$')
plot(s, F_n, linewidth = 4)
grid(b=True, which='major', color='b', linestyle='-')
xlim([0, s[-1]])
```
