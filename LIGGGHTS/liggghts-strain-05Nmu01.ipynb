```python
import pandas as pd
```

```python
df = pd.read_csv('strain15p.csv')
force = df.pressure
area = df.Area
pressure = force/area/10**6
compress = pressure[0:100]
decompress = pressure[99:201]
strain1 = linspace(0,1.5,100)
strain2 = linspace(1.5,0,102)

plot(strain1,compress,'r')
plot(strain2,decompress,'b')
legend(['Pressure','Relaxation'],4)
xlabel('strain (%)')
ylabel('stress (MPa)')
title('stress-strain diagram for 1.5% strain')
```
