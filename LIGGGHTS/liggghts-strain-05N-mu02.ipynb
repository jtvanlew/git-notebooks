```python
import pandas as pd
```

```python
df = pd.read_csv('strain05p.csv')
force = df.pressure
area = df.Area
pressure = force/area/10**6
compress = pressure[0:50]
decompress = pressure[49:101]
strain1 = linspace(0,0.5,50)
strain2 = linspace(0.5,0,52)

plot(strain1,compress,'r')
plot(strain2,decompress,'b')
legend(['Pressure','Relaxation'],4)
xlabel('strain (%)')
ylabel('stress (MPa)')
title('stress-strain diagram for 0.5% strain')
```
